```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
existing_df = pd.read_csv('trans_us.csv', index_col = 0, thousands  = ',')
existing_df.index.names = ['stations']
existing_df.columns.names = ['months']
existing_df = existing_df.fillna(15)
existing_df.head(10)
```

```python
from sklearn.cluster import KMeans
```

```python
k=KMeans(n_clusters=5)
```

```python
model=k.fit(existing_df)
```

```python
model.labels_
```

```python
outcomes=existing_df[::]
outcomes['GroupID']=model.labels_
```

```python
outcomes.head(22)
```

```python
import seaborn as sns
plt.figure(figsize=(12,12))
sns.heatmap(existing_df.corr(),annot=True)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=18)
pca.fit(existing_df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(existing_df)
```

```python
existing_2d = pca.transform(existing_df)
existing_df_2d = pd.DataFrame(existing_2d)
existing_df_2d.index = existing_df.index
existing_df_2d.columns = ['PC1','PC2']
existing_df_2d.head()
```

```python
print (pca.explained_variance_ratio_)
```

```python
sns.heatmap(existing_df_2d.corr())
```
