```python
import pandas as pd
```

```python
df=pd.read_csv('horse.csv')
```

```python
df.head(10)
```

```python
df.fillna(0).head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
surgical_label=LabelEncoder()
cp_data_label=LabelEncoder()
age_label=LabelEncoder()
main_target=LabelEncoder()
```

```python
df.columns
```

```python
df['Surgery_label']=l.fit_transform(df['surgery'])
df['Surgical_label']=surgical_label.fit_transform(df['surgical_lesion'])
df['Cp_data_label']=cp_data_label.fit_transform(df['cp_data'])
df['age_label']=age_label.fit_transform(df['age'])
df['Main_target']=main_target.fit_transform(df['outcome'])
```

```python
#df.head(10)
df[['outcome','Main_target']].head(10)
#NO means 0
#YES means 1.
```

```python
feature_n=df.drop(['surgery', 'age', 'hospital_number', 'rectal_temp', 'pulse',
       'respiratory_rate', 'temp_of_extremities', 'peripheral_pulse',
       'mucous_membrane', 'capillary_refill_time', 'pain', 'peristalsis',
       'abdominal_distention', 'nasogastric_tube', 'nasogastric_reflux',
       'nasogastric_reflux_ph', 'rectal_exam_feces', 'abdomen',
       'packed_cell_volume', 'total_protein', 'abdomo_appearance',
       'abdomo_protein', 'outcome', 'surgical_lesion', 'lesion_1', 'lesion_2',
       'lesion_3', 'cp_data', 'Encoded_label'],axis='columns')
```

```python
feature_n.head(4)
```

```python
target=df['Main_target']
target.head(10)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(feature_n,target)
```

```python
model.score(feature_n,target)
```

```python
model.predict([[0,1,0,0]])
```
